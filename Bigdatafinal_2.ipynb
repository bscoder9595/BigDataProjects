```python

stopwords_file = path_stop
stopwords_list = []

with open(stopwords_file, 'r', encoding='utf-8') as file:
    words = file.read().split()

    for word in words:
        word = word.strip()
        if word and not word.isnumeric():
            stopwords_list.append(word)
```

```python
from hazm import Normalizer, word_tokenize, stopwords_list
import re
def generate_hashtags(text):
    normalizer = Normalizer()
    normalized_text = normalizer.normalize(text)

    words = word_tokenize(normalized_text)

    words = [re.sub(r"\W+", "", w) for w in words]

    stopwords = stopwords_list()
    punctuation_marks = ['.', '،', '؛', '!', '?', '(', ')', '[', ']', '{', '}', '<', '>', '"', "'"]

    word_frequency = {}
    for word in words:
        if len(word) > 2 and word not in stopwords and word not in punctuation_marks:
            if word not in word_frequency:
                word_frequency[word] = 0
            word_frequency[word] += 1

    sorted_words = sorted(word_frequency, key=word_frequency.get, reverse=True)
    hashtag_words = sorted_words[:10]

    hashtags = [word.capitalize() for word in hashtag_words]

    return hashtags
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import udf
from pyspark.sql.types import ArrayType, StringType
from hazm import Normalizer, word_tokenize, stopwords_list
import re


generate_hashtags_udf = udf(lambda text: generate_hashtags(text), ArrayType(StringType()))

spark.udf.register("generate_hashtags_udf", generate_hashtags_udf)

hashtag_df = hashtag_df.withColumn("hashtag1", generate_hashtags_udf("article"))
```

```python
from pyspark.sql.functions import udf, explode, concat_ws
from pyspark.ml.feature import StopWordsRemover, Tokenizer
from pyspark.sql.types import StringType
from pyspark.sql.functions import desc

def extract_keywords(words):
    stopwords_list = []  
    word_frequency = {}
    for word in words:
        if word not in stopwords_list:
            if word not in word_frequency:
                word_frequency[word] = 1
            else:
                word_frequency[word] += 1
    sorted_words = sorted(word_frequency.items(), key=lambda x: x[1], reverse=True)
    keywords = [word[0] for word in sorted_words if word[0] not in stopwords_list2]
    return ", ".join(keywords[:3]) if keywords else ""

extract_keywords_udf = udf(extract_keywords, StringType())

tokenizer33 = Tokenizer(inputCol="article", outputCol="words")
df_tokenized = tokenizer33.transform(hashtag_df)

df_with_keywords = df_tokenized.withColumn("hashtag2", extract_keywords_udf("words"))

df_exploded = df_with_keywords.select("article", "hashtag2")

df_final = df_exploded.join(hashtag_df, "article")

df_final.show(truncate=False)
```

```python
def filter_stopwords(word_list):
    words = [word.strip() for word in word_list if word.strip() not in stopwords_list2]
    return ', '.join(words[:5])

# Apply the function to the 'hashtag1' column
df_f_p['hashtag1_filtered'] = df_f_p['hashtag1'].apply(filter_stopwords)
```
